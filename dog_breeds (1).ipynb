```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import string
import time
```

```python
alphabets = list(string.ascii_uppercase)
```

```python
breed_names = []
for i in alphabets:
    response = requests.get("http://www.akc.org/dog-breeds/?letter=" + i)
    doc = BeautifulSoup(response.text, "html.parser")
    list_of_dogs = doc.find_all('h2')[1:]
    breed_names += list_of_dogs
```

```python
breed_names_clean = []
for i in breed_names:
    breed_names_clean.append(i.text.replace(' ', '-').replace(' ', '').replace('.', '-'))
```

```python
df = pd.DataFrame(columns=['name',
                  'personality',
                  'energy',
                  'height'])

for i in breed_names_clean:
    print(i)
    time.sleep(1)
    try:
        response = requests.get("http://www.akc.org/dog-breeds/" + i)
        doc = BeautifulSoup(response.text, "html.parser")
        name = doc.find('h1').text
        personality = doc.find_all('strong')[0].next_sibling
        energy = doc.find_all('strong')[1].next_sibling
        height = doc.find_all('strong')[7].next_sibling
        data = pd.Series({'name': name, 
                          'personality': personality, 
                          'energy': energy,
                          'height': height
                         })
        df = df.append(data, ignore_index=True)
    except:
        print("failure on breed:", i)
```
